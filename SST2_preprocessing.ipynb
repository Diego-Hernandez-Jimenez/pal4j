```python
from transformers import pipeline
from datasets import load_dataset
import numpy as np
import pandas as pd
```

```python
ds = load_dataset("stanfordnlp/sst2")
```

```python
feature_extractor = pipeline("feature-extraction", model='distilbert-base-uncased')

extracted_features = []
labels = []
for instance in ds["train"]:
    features = feature_extractor(instance["sentence"], return_tensors="pt")[0]
    # pool the features
    reduced_features = features.numpy().mean(axis=0)
    extracted_features.append(reduced_features)
    labels.append(instance["label"])

features_df = pd.DataFrame(extracted_features)
labels_series = pd.Series(labels, name='label')
combined_df = pd.concat([features_df, labels_series], axis=1)

combined_df.columns = [f"x{i}" for i in range(combined_df.shape[1] - 1)] + ['label']
combined_df["label"] = combined_df["label"].apply(lambda x: -1 if x == 0 else 1) # change label
combined_df

print("DataFrame created with features and labels:")
combined_df.head()
```
